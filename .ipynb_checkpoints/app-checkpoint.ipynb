```python
# 1. Imports
import pandas as pd
import matplotlib.pyplot as plt
from utils import match_player_name, estimate_player_worth

#  2. Load the Data
deliveries = pd.read_csv("deliveries.csv")
matches = pd.read_csv("matches.csv")

#  3. Add 'season' to deliveries using match_id from matches
match_season_map = matches.set_index("id")["season"].to_dict()
deliveries["season"] = deliveries["match_id"].map(match_season_map)

#  4. Define Stats Function
def get_player_stats(player_name, deliveries_df):
    player_deliveries = deliveries_df[deliveries_df["batter"] == player_name]

    total_runs = player_deliveries["batsman_runs"].sum()
    balls_faced = player_deliveries.shape[0]
    strike_rate = (total_runs / balls_faced) * 100 if balls_faced > 0 else 0
    num_4s = player_deliveries[player_deliveries["batsman_runs"] == 4].shape[0]
    num_6s = player_deliveries[player_deliveries["batsman_runs"] == 6].shape[0]

    return {
        "Player": player_name,
        "Total Runs": total_runs,
        "Balls Faced": balls_faced,
        "Strike Rate": round(strike_rate, 2),
        "4s": num_4s,
        "6s": num_6s
    }

#  5. User Input & Fuzzy Match
input_name = input("Enter player name: ").strip().lower()
all_players = deliveries["batter"].unique().tolist()
matched_name = match_player_name(input_name, all_players)

if not matched_name:
    print("âŒ Player not found.")
else:
    print(f"\nâœ… Matched Player: {matched_name}")

    #  6. Show Stats
    stats = get_player_stats(matched_name, deliveries)
    print("\n--- Player Stats ---")
    for k, v in stats.items():
        print(f"{k}: {v}")

    #  7. Worth Estimation
    worth = estimate_player_worth(stats)
    print("\n--- Worth Estimation ---")
    for k, v in worth.items():
        print(f"{k}: {v}")

    #  8. Graphs - 4s & 6s by Season
    fours_sixes = deliveries[(deliveries["batter"] == matched_name) &
                             (deliveries["batsman_runs"].isin([4, 6]))]
    fs_count = fours_sixes.groupby(["season", "batsman_runs"]).size().unstack().fillna(0)
    fs_count.plot(kind="bar", stacked=True, color=["green", "red"])
    plt.title(f"{matched_name} - 4s & 6s by Season")
    plt.ylabel("Count")
    plt.xlabel("Season")
    plt.legend(["4s", "6s"])
    plt.tight_layout()
    plt.show()

    #  9. Graph - Strike Rate by Season
    player_season_data = deliveries[deliveries["batter"] == matched_name].copy()
    season_stats = player_season_data.groupby("season").agg({
        "batsman_runs": "sum",
        "ball": "count"
    })
    season_stats["strike_rate"] = (season_stats["batsman_runs"] / season_stats["ball"]) * 100
    season_stats["strike_rate"].plot(marker='o', color='blue')
    plt.title(f"{matched_name} - Strike Rate by Season")
    plt.xlabel("Season")
    plt.ylabel("Strike Rate")
    plt.grid(True)
    plt.show()

    #  10. Insights
    peak_season = season_stats["batsman_runs"].idxmax()
    print(f"ðŸ”¥ Peak Season: {peak_season} with {season_stats.loc[peak_season]['batsman_runs']} runs")

    consistent_years = (season_stats["batsman_runs"] > 300).sum()
    print(f"ðŸ“ˆ Consistent Seasons (300+ runs): {consistent_years}")
```
